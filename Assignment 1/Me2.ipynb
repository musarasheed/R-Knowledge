```python
import math
def fac(x):
    return math.factorial(x)

outputs = []
nums = [92,93,94,95]
p = 0.933333333333333333333333333333333333333333333333333333
n = 95

for i in nums:
    a = fac(n)/(fac(i) * fac(n-i)) #Factorial 
    
    b = p**(i) #(p)^x
    
    c = (1-p)**(n-i) #(1-p)^(n-x)
    
    outputs.append(a*b*c)

sum(outputs)
```
